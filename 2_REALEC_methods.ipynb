```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl
```

## Preprocessing

```python
data['_name'] = data['name'].apply(lambda r: r.split('_')[-1])
```

```python
groups = data.groupby('_name')
frst = groups.get_group('1.txt').drop(['_name'], axis=1)
scnd = groups.get_group('2.txt').drop(['_name'], axis=1)
```

## Multi Layer Perceptron Classifier

```python
scaler = StandardScaler()
fof = scaler.fit_transform(frst.iloc[:, 2:])
X_train, X_test, y_train, y_test = train_test_split(fof, frst['CEFR'], test_size=0.25)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.exceptions import ConvergenceWarning
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split



mlp = MLPClassifier(
    hidden_layer_sizes=(1000,)
    ,activation='relu',
    solver='adam',
    alpha=0.1, batch_size='auto',
    learning_rate='constant', learning_rate_init=0.004,
    power_t=0.5, max_iter=200, shuffle=True, random_state=None,
    tol=0.0001, verbose=False, warm_start=True, momentum=0.9, 
    nesterovs_momentum=True, early_stopping=False,
    validation_fraction=0.1, beta_1=0.9, beta_2=0.999, 
    epsilon=1e-08, n_iter_no_change=10, max_fun=15000
)
mlp.fit(X_train, y_train)

print("Training set score: %f" % mlp.score(X_train, y_train))
print("Test set score: %f" % mlp.score(X_test, y_test))
```

## Extra Trees Classifier

```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_blobs
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier


clf = ExtraTreesClassifier(n_estimators=1000, max_depth=None,
    min_samples_split=2, random_state=0)
scores = cross_val_score(clf, fof, frst['CEFR'], cv=5)


```

## Histogram-based Gradient Boosting Classification Tree

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.ensemble import VotingClassifier

clf = HistGradientBoostingClassifier(max_iter=100)
scores = cross_val_score(clf, fof, frst['CEFR'], cv=5)
scores.mean()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

clf1 = LogisticRegression(random_state=1)
clf2 = RandomForestClassifier(n_estimators=50, random_state=1)
clf3 = GaussianNB()

eclf = VotingClassifier(
    estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)],
    voting='soft')

scores = cross_val_score(eclf, fof, frst['CEFR'], cv=5)
scores.mean()
```

## CatBoost Classification

```python
import catboost
from catboost import CatBoostClassifier
```

```python
clf = CatBoostClassifier()
```

```python
def classify(df):
    X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 2:], df['CEFR'], test_size=0.25)
    clf.fit(X_train, y_train, eval_set=(X_test, y_test))
    return clf.score(X_test, y_test)
```
