```python
def read_txt(name, path='/home/daria/REALEC/NEW_DATA/cefr_data_texts/'):
    with open(Path(path) / name) as f:
        text = f.read()
    return text
        
```

```python
def length(text):
    return len(text.split(' '))
```

```python
nw1 = []

for name in list(data1['name']):
    nw1.append(length(read_txt(name)))
```

```python
nw2 = []

for name in list(data2['name']):
    nw2.append(length(read_txt(name)))
```

```python
data1.insert(1, 'num_of_words', nw1)
```

```python
clf = CatBoostClassifier()
```

```python
def classify(df):
    X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 3:], df['CEFR_labelled'], test_size=0.2, random_state=42)
    clf.fit(X_train, y_train, eval_set=(X_test, y_test))
    return clf.score(X_test, y_test)
```

```python
classify(data1)
```

```python
classify(data2)
```
